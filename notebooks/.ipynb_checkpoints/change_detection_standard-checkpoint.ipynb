```python
# Note: this may well go into a plot_funcs script.
# This must be updated, taking out the I/O and just 
# passing the relevant parameters
def make_rgb(ztstart, ztend):
    """
    Helper function to make RGB-like images
    from the visible cubes
    """
    wl_rgb = [640., 560., 480.]
    yy, xx, vwl, _, vrf, _, info = read_prs_l2d(path_rawdata, ztstart, ztend)
    img = vrf.astype(float)

    # Get indices for the RGB channels
    idx_rgb = []
    for wl in wl_rgb:
        idx_rgb.append(np.abs(vwl-wl).argmin())

    rgb = img[:,:, idx_rgb]

    # Normalize frames for better visualization. 
    # Using np.max() is not good if there are clouds or bright features.
    for iw in range(len(wl_rgb)):
        rgb[:,:,iw] /= np.percentile(rgb[:,:,iw],99.5)
    
    return rgb
```

Here follow the standard, single or multi band difference methods

```python
bsel = 1000.
```

```python
# Standard difference
sdif = diffs(wvl_mv, boa_tg, boa_mv, bsel=bsel, mode='sdif')
```

```python
# Log-ratio
logr = diffs(wvl_mv, boa_tg, boa_mv, bsel=bsel, mode='logr')
```

```python
# Generalized likelihood ratio
glrt = diffs(wvl_mv, boa_tg, boa_mv, bsel=bsel, mode='glrt')
```

```python
# Cosine similarity
coss = diffs(wvl_mv, boa_tg, boa_mv, mode='coss')
```

```python
plt.figure(figsize=(20,14))
plt.subplot(2,2,1)
plt.imshow(np.abs(sdif))
plt.colorbar(orientation='vertical')
plt.subplot(2,2,2)
plt.imshow(logr)
plt.colorbar(orientation='vertical')
plt.subplot(2,2,3,)
plt.imshow(glrt)
plt.colorbar(orientation='vertical')
plt.subplot(2,2,4,)
plt.imshow(coss)
plt.colorbar(orientation='vertical')
```
