```python
# Set to True to show plots, save them, and/or save processed files
make_plots = True
save_plots = False
save_files = False
```

```python
# May be good to move these to a file at the same level of config, to just change there...
tstart1 = '20200115095908'; tend1 = '20200115095912'
tstart2 = '20210713095222'; tend2 = '20210713095226'

y1, x1, vwl1, swl1, vrf1, srf1, info1 = read_prs_l2d(path_rawdata, tstart1, tend1, units=True)
y2, x2, vwl2, swl2, vrf2, srf2, info2 = read_prs_l2d(path_rawdata, tstart2, tend2, units=True)
print(np.min(vrf1), np.min(vrf2), np.max(vrf1), np.max(vrf2))
print(np.min(srf1), np.min(srf2), np.max(srf1), np.max(srf2))
```

```python
# Check if wavelengths are the same witin a tolerance
if np.max(np.abs(vwl1-vwl2)) > 2.:
    raise ValueError('Different VNIR bands!')
if np.max(np.abs(swl1-swl2)) > 2.:
    raise ValueError('Different SWIR bands!')
```

```python
# Note the shapes can be different. Here just reading the VIS data.
print('Grid 1', x1.shape, y1.shape)
print('Grid 2', x2.shape, y2.shape)
```

```python
# Show how coordinates are organized in the L2D files (different in other cases)
if make_plots is True:
    pad = 0.05
    stride = 1
    if x1.shape[0] > 900:
        stride = 200

    plt.rcParams['figure.figsize'] = [16, 8]
    fig, (ax1,ax2) = plt.subplots(1,2)

    for i1 in range(len(x1[:,0]))[::stride]:
        for j1 in range(len(y1[0,:]))[::stride]:
            ax1.scatter(x1[i1,j1], y1[i1,j1], s=1, c='r',marker='o')
            ax1.annotate('{},{}'.format(i1,j1), (x1[i1,j1], y1[i1,j1]),
                         color='r',fontsize=10)#rotation=-45)
    ax1.set_xlim([np.min(x1)-pad, np.max(x1)+pad])
    ax1.set_ylim([np.min(y1)-pad, np.max(y1)+pad])

    for i2 in range(len(x2[:,0]))[::stride]:
        for j2 in range(len(y2[0,:]))[::stride]:
            ax2.scatter(x2[i2,j2], y2[i2,j2], s=1, c='b',marker='o')
            ax2.annotate('{},{}'.format(i2,j2), (x2[i2,j2], y2[i2,j2]),
                         color='b', fontsize=10)
    ax2.set_xlim([np.min(x2)-pad, np.max(x2)+pad])
    ax2.set_ylim([np.min(y2)-pad, np.max(y2)+pad])

    plt.tight_layout()
```

```python
# Determine the intersection of the two grids from their corners
xnw, ynw, xse, yse = find_corners(x1, x2, y1, y2)
```

```python
# Show all grid points of the two files
if make_plots is True:
    plt.figure(figsize=(10,10), dpi=100) # 600
    rstride = int(stride/5)
    plt.scatter(x1[::rstride, ::rstride], y1[::rstride, ::rstride], 
                s=10, c='r', marker='.', label='G1')
    plt.scatter(x2[::rstride, ::rstride], y2[::rstride, ::rstride], 
                s=10, c='b', marker='.', label='G2')
    plt.scatter(xse, yse, s=20, c='g', marker='x', label='SE')
    plt.scatter(xnw, ynw, s=20, c='g', marker='*', label='NW')
    plt.legend(ncol=4)
```

```python
# Zoom over the corner regions. This is very slow with stride set to 1.
if make_plots is True:
    plt.figure(figsize=(10,5), dpi=10)# 600
    rstride = 1
    pad = 0.002

    plt.subplot(1,2,1)
    plt.scatter(x1[::rstride, ::rstride], y1[::rstride, ::rstride], 
                s=10, c='r', marker='.', label='G1')
    plt.scatter(x2[::rstride, ::rstride], y2[::rstride, ::rstride], 
                s=10, c='b', marker='.', label='G2')
    plt.scatter(xse, yse, s=20, c='g', marker='x', label='SE')
    plt.scatter(xnw, ynw, s=20, c='g', marker='*', label='NW')
    plt.xlim([xse-pad, xse+pad])
    plt.ylim([yse-pad, yse+pad])
    plt.legend(ncol=2)

    plt.subplot(1,2,2)
    plt.scatter(x1[::rstride, ::rstride], y1[::rstride, ::rstride], 
                s=10, c='r', marker='.', label='G1')
    plt.scatter(x2[::rstride, ::rstride], y2[::rstride, ::rstride], 
                s=10, c='b', marker='.', label='G2')
    plt.scatter(xse, yse, s=20, c='g', marker='x', label='SE')
    plt.scatter(xnw, ynw, s=20, c='g', marker='*', label='NW')
    plt.xlim([xnw-pad, xnw+pad])
    plt.ylim([ynw-pad, ynw+pad])
    plt.legend(ncol=2)
```

```python
# Find corner indices in grids
tol = 30 # m
i1nw, j1nw = idx_mesh(xnw, ynw, x1, y1, maxdist=tol)
i2nw, j2nw = idx_mesh(xnw, ynw, x2, y2, maxdist=tol)
i1se, j1se = idx_mesh(xse, yse, x1, y1, maxdist=tol)
i2se, j2se = idx_mesh(xse, yse, x2, y2, maxdist=tol)

# Check that NW/SE ordering is consistent
if i1nw > i1se or j1nw > j1se: 
    raise ValueError('Unexpected coordinate order')

if i2nw > i2se or j2nw > j2se: 
    raise ValueError('Unexpected coordinate order')

# Print coordinates for inspection
print('')
print('Crd NW: {} {}'.format(xnw, ynw))
print('Grid 1:', x1[i1nw, j1nw], y1[i1nw, j1nw])
print('Grid 2:', x2[i2nw, j2nw], y2[i2nw, j2nw])
print('')
print('Crd SE: {} {}'.format(xse, yse))
print('Grid 1:', x1[i1se, j1se], y1[i1se, j1se])
print('Grid 2:', x2[i2se, j2se], y2[i2se, j2se])
```

```python
# Save cropped images, conventionally assuming the first as 'target' 
# (reference) and the second as 'moving'. The order has no particular meaning.

# Use first file as target
intx = x1[i1nw:i1se+1, j1nw:j1se+1]
inty = y1[i1nw:i1se+1, j1nw:j1se+1]

# Define 4 cropped datacubes for interpolation 
# VIS
iimg1_v = np.zeros(np.shape(intx)+(len(vwl1),))
iimg2_v = np.zeros(np.shape(intx)+(len(vwl1),))
# SWIR
iimg1_s = np.zeros(np.shape(intx)+(len(swl1),))
iimg2_s = np.zeros(np.shape(intx)+(len(swl1),))
print(np.shape(iimg1_v),np.shape(iimg2_v))
```

```python
# Populate overlapping datacubes, note zeros are in the rest
# of the datacube. Maybe loop can be vectorized?
for ii in range(np.shape(intx)[0]-1):
    for jj in range(np.shape(intx)[1]-1):
        iimg1_v[ii, jj, :] = vrf1[i1nw+ii, j1nw+jj, :]
        iimg2_v[ii, jj, :] = vrf2[i2nw+ii, j2nw+jj, :]
        iimg1_s[ii, jj, :] = srf1[i1nw+ii, j1nw+jj, :]
        iimg2_s[ii, jj, :] = srf2[i2nw+ii, j2nw+jj, :]
print(np.shape(iimg1_v),np.shape(iimg2_s))

# Conform distribution of pixels equal 0
for iwl, wl in enumerate(vwl1):
    iimg1_v[:,:,iwl][iimg2_v[:,:,iwl] == 0] = 0.
    iimg1_s[:,:,iwl][iimg2_v[:,:,iwl] == 0] = 0.
    iimg2_v[:,:,iwl][iimg1_v[:,:,iwl] == 0] = 0.
    iimg2_s[:,:,iwl][iimg1_v[:,:,iwl] == 0] = 0.
```

```python
if make_plots is True:
    plt.figure(figsize=[20,10])
    plt.subplot(1,2,1)
    plt.imshow(iimg1_v[:,:,0]/np.max(iimg1_v[:,:,0]),cmap='Greys')
    plt.title("2019 - target masked")

    plt.subplot(1,2,2)
    plt.imshow(iimg2_v[:,:,0]/np.max(iimg2_v[:,:,0]),cmap='Greys')
    plt.title("2020 - moving masked")
```

```python
# Call gefolki to compute the optical flow
u_of, v_of = EFolki(iimg1_v[:,:,0], iimg2_v[:,:,0], iteration=4, radius=[5,5], rank=4, levels=5)
```

```python
print(u_of.shape, v_of.shape)
```

```python
if make_plots is True:
    plt.figure(figsize=[20,6])
    #plt.subplot(1,3,1)
    plt.imshow(np.abs(u_of)+np.abs(v_of),vmin=0,vmax=4)
    xx, yy = np.meshgrid(np.arange(u_of.shape[0]),np.arange(u_of.shape[1]))
    # Quiver does not look good since there are too many points.
    # Streamplot automatically adjust density.
    plt.streamplot(np.arange(u_of.shape[1]), np.arange(u_of.shape[0]), u_of, v_of, 
                   density=3, color='r', linewidth=0.5)
    plt.title('Norm of Optic to optic registration - Red')
    plt.xlim([0, u_of.shape[1]]) # as in plots above
    plt.ylim([u_of.shape[0], 0])
    plt.colorbar()
```

```python
# Coregister the moving datacube with the optical flow
iimg2_vc = np.zeros(np.shape(intx)+(len(vwl1),))
iimg2_sc = np.zeros(np.shape(intx)+(len(vwl1),))

for iwl,band in enumerate(vwl2):
        iimg2_vc[:,:,iwl] = wrapData(iimg2_v[:,:,iwl], u_of, v_of)
        iimg2_sc[:,:,iwl] = wrapData(iimg2_s[:,:,iwl], u_of, v_of)
```

```python
if make_plots is True:
    plt.figure(figsize=[20,10])
    plt.subplot(1,2,1)
    plt.imshow(iimg1_v[:,:,0]/np.max(iimg1_v[:,:,0]),cmap='Greys')
    plt.title("2019 - target orig")

    plt.subplot(1,2,2)
    plt.imshow(iimg2_vc[:,:,0]/np.max(iimg2_vc[:,:,0]),cmap='Greys')
    plt.title("2020 - moving coreg")
```
